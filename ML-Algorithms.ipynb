```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('kidney_disease.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
#Renaming The Columns

df.rename(columns={'bp': 'blood_pressure', 'sg': 'specific gravity','al':'albumin','su':'sugar','rbc':'red blood cells','pc':'pus cell','pcc':'pus cell clumps','ba':'bacteria','bgr':'blood glucose random','bu':'blood urea','sc':'serum creatinine','sod':'sodium','pot':'potassium','hemo':'hemoglobin','pcv':'packed cell volume','wc':'white blood cell count','rc':'red blood cell count','htn':'hypertension','dm':'diabetes mellitus','cad':'coronary artery disease','appet':'appetite','pe':'pedal edema','ane':'anemia','classification':'class'},inplace=True)
```

```python
df.head(10)
```

```python
#Dropping the id colunm

df.drop('id', axis = 1, inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
#Converting some necessary columns to the right data type

df['packed cell volume'] = pd.to_numeric(df['packed cell volume'], errors='coerce')
df['white blood cell count'] = pd.to_numeric(df['white blood cell count'], errors='coerce')
df['red blood cell count'] = pd.to_numeric(df['red blood cell count'], errors='coerce')
df['age'] = df['age'].astype('int')
```

```python
df.info()
```

```python
#Checking for null values

df.isnull().sum()
```

```python
#filing  the numerical columns with the mean

df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
#Filling the categorical column with the most common class

df_clean = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df_clean.head(10)
```

```python
df_clean.isnull().sum()
```

```python
#Extracting the numerical and categorical columns

cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
#Checking the distribution of the numerical features

plt.figure(figsize = (20, 15))
plotnumber = 1

for column in num_cols:
    if plotnumber <= 14:
        ax = plt.subplot(3, 5, plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```

#### As seen in the distribution plot, we can see some level of skewness in the colunms.

```python
#Glancing through the categorical variable

plt.figure(figsize = (20, 15))
plotnumber = 1

for column in cat_cols:
    if plotnumber <= 11:
        ax = plt.subplot(3, 4, plotnumber)
        sns.countplot(df[column], palette = 'rocket')
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
#Checking for correlation using the heatmap

plt.figure(figsize = (15, 8))

sns.heatmap(df.corr(), annot = True, linewidths = 2, linecolor = 'lightgrey')
plt.show()
```

```python
#Encoding the categorical features using label Encoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

## Lets Build Our Models

### Logistic Regression

```python
#Features Extraction

X = df.loc[:, ['specific gravity', 'albumin', 'blood glucose random', 'hemoglobin', 'packed cell volume','red blood cell count', 'hypertension']]
y = df['class'].values
```

```python
model.predict(X_test)
```

```python
print('Accuracy={:.2f}'.format(model.score(X_test, y_test)))
```

```python
model.score(X_train, y_train)
```

```python
#Confusion Matrix

y_pred = model.predict(X_test)

confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

#### 88 True Positive: Actual Positive and Predicted as Positive
####  51 True Negative: Actual Negative and Predicted as Negative
#### 1 False Positive(Type I Error): Actual Negative but predicted as Positive
#### 0 False Negative(Type II Error): Actual Positive but predicted as Negative

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

## Decision Trees

```python
#Extracting the features

X = df.loc[:, ['specific gravity', 'albumin', 'blood glucose random', 'hemoglobin', 'packed cell volume','red blood cell count', 'hypertension']]
y = df['class'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30,random_state=20)

model = DecisionTreeClassifier(criterion = 'gini', splitter = 'random', max_leaf_nodes = 15, min_samples_leaf = 5, max_depth = 5)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy_score(y_pred, y_test)
```

```python
accuracy_score(model.predict(X_train), y_train)
```

```python
import graphviz
graph_ = tree.export_graphviz(model, label='all', rounded=True, filled=True)
graph = graphviz.Source(graph_)

graph.render('data')
graph
```

## Random Forest

```python
#Extracting the features

X = df.loc[:, ['specific gravity', 'albumin', 'blood glucose random', 'hemoglobin', 'packed cell volume','red blood cell count', 'hypertension']]
y = df['class'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier #Importing Random Forest Classifier
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=20)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(f'The accuracy for the test data : {metrics.accuracy_score(y_pred, y_test)}')
```

```python
y_pred_train = model.predict(X_train)
print(f'The accuracy for the training data : {metrics.accuracy_score(y_pred_train, y_train)}')
```

```python
from sklearn.model_selection import cross_val_score

model = RandomForestClassifier(n_estimators=100)
scores = cross_val_score(model, X, y, cv=5)

print(f'Accuracy on each fold : {scores}')

print(f'Average of the accuracy :{sum(scores)/5}')
```

```python
from sklearn.metrics import silhouette_score
print(f'Silhouette Score(n=2):{silhouette_score(X_train, model.predict(X_train))}')
```

```python
#Getting the optimal K

sum_of_squared_dist = []

for i in range(1,10):
  model = KMeans(n_clusters=i)
  model.fit(X_train)
  sum_of_squared_dist.append(model.inertia_)
```

```python
plt.figure(figsize=(16,8))

plt.plot(range(1,10), sum_of_squared_dist, color='r')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared error')
plt.title('Sum of squared error vs Number of cluster')
plt.show()
```
